```python
a = {1, 2, 3}
b = {3, 4, 5}

# union
print("union")
print(a | b)           # {1, 2, 3, 4, 5}
print(a.union(b))      # 同上
print("-----------------")
# intersection
print("intersection")
print(a & b)           # {3}
print(a.intersection(b))
print("-----------------")
# difference
print("difference")
print(a - b)           # {1, 2}
print(a.difference(b))
print("-----------------")
# symmetric difference (XOR)
print("symmetric difference (XOR)")
print(a ^ b)           # {1, 2, 4, 5}
print(a.symmetric_difference(b))
```

# map() filer() reduce()

```python
# map() on list
L = list(range(1,21))
D = list(map(lambda x: x**2, L))
print(D)

```

```python
# map() on dictionary
students = {"Alice": 85, "Bob": 58, "Charlie": 92}

curved = dict(map(lambda item: (item[0], item[1] + 5), students.items()))
print(curved)

```

```python
# filter() on list
L = list(range(1,21))
even_numbers = list(filter(lambda x: x%2==0, L))
print(even_numbers)
```

```python
# https://www.geeksforgeeks.org/python/reduce-in-python/
from functools import reduce
li = ["Geeks", "for", "Geeks"]
res = reduce(lambda x, y: x + " " + y, li)
print(res)
```

```python
# filter() on dictionary
students = {
    "Alice": 85,
    "Bob": 58,
    "Charlie": 92,
    "David": 45
}

passed = dict(filter(lambda item: item[1] >= 60, students.items()))
print(passed)

```

```python
# reduce() on list
from functools import reduce
L = [1,2,3,4,5]
product = reduce(lambda x, y: x * y, L,1)
print(product)
```

```python
# find the maximum value in a list
nums = [3, 8, 2, 10, 5]
max_num = reduce(lambda a, b: a if a > b else b, nums)
print(max_num)
```

```python
from functools import reduce

students = {"Alice": 85, "Bob": 58, "Charlie": 92}

total = reduce(lambda acc, item: acc + item[1], students.items(),0) # The first element is a tuple which cant be added directly
print(total)  
```

# any() all()

```python
print(any([True, False, True]))
print(any([0, False, 1]))

print(all([True, True, True]))
print(all([True, False, True]))
print(all([1, 2, 3, 0]))
```

```python
students = {
    "Alice": 85,
    "Bob": 58,
    "Charlie": 92,
    "David": 45
}
print(all(score >= 60 for score in students.values()))
print(any(score < 60 for score in students.values()))
```

# some basic data structures 


```python
from queue import Queue
q = Queue()
for i in range(5):
    q.put(i)
    print(f"Enqueued: {i}")

while not q.empty():
    print(f"Dequeued: {q.get()}")
# Stack can be implemented using list
```

```python
from collections import deque
dq = deque([1,2,3])
dq.append(4)        
dq.appendleft(0)
print(dq)
print("first element:",dq[0])
print("last element:",dq[-1])
dq.pop()
print("after pop():",dq)
dq.popleft()
print("after popleft():",dq)
dq.rotate(1)
print("after rotate(1):",dq)
dq.rotate(-2)
print("after rotate(-2):",dq)

```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-2*np.pi, 2*np.pi, 1000)
y = np.sin(x)
z = np.cos(x)
plt.plot(x, y)
plt.title("Sine and Cosine Functions")
plt.xlabel("x")
plt.ylabel("sin(x) and cos(x)")
plt.plot(x, z)
plt.legend(["sin(x)","cos(x)"])
```
